## Histogram


```python
from utils import *
%matplotlib inline
# Define date range
start_date ='2009-01-01'
end_date = '2012-12-31' 
dates = pd.date_range(start_date,end_date)
symbols = ['SPY','XOM']
data = get_data(symbols,dates)
plot_data(data)

daily_returns = compute_daily_returns(data)
plot_data(daily_returns, title="Daily Returns", y_label="Daily returns")

# Plot Histogram
daily_returns['SPY'].hist(bins=20)
mean = daily_returns['SPY'].mean()

std = daily_returns['SPY'].std()
kurtosis = daily_returns['SPY'].kurtosis()
print("Mean: {}, Std: {}, Kurtosis: {}".format(mean, std, kurtosis))

# Plot the mean and std line
plt.axvline(mean, color='w', linestyle='dashed', linewidth=2)
plt.axvline(std, color='r', linestyle='dashed', linewidth=2)
plt.axvline(-std, color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
daily_returns['SPY'].hist(bins=20, label='SPY')
daily_returns['XOM'].hist(bins=20, label='XOM')
plt.legend(loc='upper left')
plt.show()
```

## Scatter

```python
import numpy as np

# Define date range
start_date ='2009-01-01'
end_date = '2012-12-31' 
dates = pd.date_range(start_date,end_date)
symbols = ['SPY','XOM','GLD']
data = get_data(symbols,dates)
plot_data(data)

daily_returns = compute_daily_returns(data)

# Scatterplot SPY vs XOM
daily_returns.plot(kind='scatter',x='SPY', y='XOM')
coefficient = np.polyfit(daily_returns['SPY'],daily_returns['XOM'],1)
beta_XOM,alpha_XOM = coefficient[0], coefficient[1]  
print("Beta_XOM: {}, Alpha_XOM: {}".format(beta_XOM,alpha_XOM))
plt.plot(daily_returns['SPY'], daily_returns['SPY']*beta_XOM+alpha_XOM,'-',color='r')
plt.show()

# Scatterplot SPY vs GLD
daily_returns.plot(kind='scatter',x='SPY', y='GLD')
coefficient = np.polyfit(daily_returns['SPY'],daily_returns['GLD'],1)
beta_GLD, alpha_GLD = coefficient[0], coefficient[1] 
print("Beta_GLD: {}, Alpha_GLD: {}".format(beta_GLD,alpha_GLD))
plt.plot(daily_returns['SPY'], daily_returns['SPY']*beta_GLD+alpha_GLD,'-',color='r')
plt.show()
```
