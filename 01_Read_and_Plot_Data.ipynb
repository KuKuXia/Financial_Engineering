## Read Data
注意：“Adj Close”是“调整后的收盘价”的简称，即考虑企业行为（例如 stock splits）和分布（例如 dividends issued）的历史调整股价。

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
df = pd.read_csv("Data/AAPL.csv")
```

## Find the Highest Price

```python
def get_max_close(symbol):
    df = pd.read_csv("Data/{}.csv".format(symbol))
    return df['Close'].max()
for symbol in ["AAPL", "IBM"]:
    print("Max Close: ")
    print(get_max_close(symbol))
```

## Find the Mean Volume

```python
def get_max_close(symbol):
    df = pd.read_csv("Data/{}.csv".format(symbol))
    return df['Volume'].mean()
for symbol in ["AAPL", "IBM"]:
    print("Mean Volume: ")
    print(get_max_close(symbol))
```

## Plot Adj Close Data

```python
def plot_data(symbol):
    df = pd.read_csv("Data/{}.csv".format(symbol))
    df = df[-3500:]
    df['Adj Close'].plot()
    plt.title("Adj Close Price of {}".format(symbol))
    plt.show()
for symbol in ["AAPL", "IBM"]:
    print("Plot Data: ")
    print(plot_data(symbol))
```

## Plot High Price Data

```python
def plot_data(symbol):
    df = pd.read_csv("Data/{}.csv".format(symbol))
    df = df[-3500:]
    df['High'].plot()
    plt.title("High Price of {}".format(symbol))
    plt.show()
for symbol in ["AAPL", "IBM"]:
    print("Plot High Data: ")
    print(plot_data(symbol))
```

## Plot Two Columns 

```python
def plot_data(symbol):
    df = pd.read_csv("Data/{}.csv".format(symbol))
    df = df[-3500:]
    df[['Close','Adj Close']].plot()
    plt.title("Close and Adj Close Price of {}".format(symbol))
    plt.show()
for symbol in ["AAPL", "IBM"]:
    print("Plot Data: ")
    print(plot_data(symbol))
```
